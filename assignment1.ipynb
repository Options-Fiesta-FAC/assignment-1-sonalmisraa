```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_payoffs(K, P, S_range):
    #to calculate the payoffs for all option strategies
    buy_call = np.maximum(0, S_range-K) - P #buying a call option
    sell_call = P - np.maximum(0,S_range -K) #selling a call option
    buy_put = np.maximum(0,K-S_range) - P # buying a put option
    sell_put = P - np.maximum(0,K-S_range) #selling a put option
    return buy_call,sell_call,buy_put,sell_put

def plot_payoffs(S_range, payoffs, K, P): # to plot the four payoff diagrams 
    
    buy_call, sell_call, buy_put, sell_put = payoffs
    plt.figure(figsize=(14, 10))

    # Create subplots
    plt.subplot(2, 2, 1)
    plt.plot(S_range, buy_call, 'b', label='Buy Call')
    plt.title('Buy Call Option')
    plt.xlabel('Spot Price (S)')
    plt.ylabel('Profit/Loss')
    plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
    plt.axvline(K, color='blue', linestyle=':', linewidth=0.5, label='Strike Price')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(2, 2, 2)
    plt.plot(S_range, sell_call, 'r', label='Sell Call')
    plt.title('Sell Call Option')
    plt.xlabel('Spot Price (S)')
    plt.ylabel('Profit/Loss')
    plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
    plt.axvline(K, color='blue', linestyle=':', linewidth=0.5, label='Strike Price')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(2, 2, 3)
    plt.plot(S_range, buy_put, 'g', label='Buy Put')
    plt.title('Buy Put Option')
    plt.xlabel('Spot Price (S)')
    plt.ylabel('Profit/Loss')
    plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
    plt.axvline(K, color='blue', linestyle=':', linewidth=0.5, label='Strike Price')
    plt.legend()
    plt.grid(True)
    
    plt.subplot(2, 2, 4)
    plt.plot(S_range, sell_put, 'm', label='Sell Put')
    plt.title('Sell Put Option')
    plt.xlabel('Spot Price (S)')
    plt.ylabel('Profit/Loss')
    plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
    plt.axvline(K, color='blue', linestyle=':', linewidth=0.5, label='Strike Price')
    plt.legend()
    plt.grid(True)
    
    plt.suptitle(f'Option Payoff Diagrams (Strike Price = {K}, Premium = {P})')
    plt.tight_layout()
    plt.show()

def main():     # Get user inputs
    try:
        K = float(input("Enter the strike price (K): "))
        P = float(input("Enter the premium (P): "))
    except ValueError:
        print("Please enter valid numbers for strike price and premium.")
        return
    
    
    S_range = np.linspace(K - 50, K + 50, 500) # Create a range of spot prices (K-50 to K+50)
    
   
    payoffs = calculate_payoffs(K, P, S_range)  # Calculate all payoffs
    
  
    plot_payoffs(S_range, payoffs, K, P)   # Plot the results

if __name__ == "__main__":
    main()  
```

```python
"""
Buying a Call Option:Profit is unlimited as the spot price increases above the strike price.
                     Maximum loss is limited to the premium paid.
                     Breakeven point :the buyer pays a premium upfront, so the stock must rise enough to recover this cost. The option only has value when the spot price (S) exceeds K, but the profit must also cover P. Thus, the breakeven is where 
                     S=K+P.

Selling a Call Option :Profit is limited to the premium received
                       Loss is unlimited as the spot price increases above the strike price
                       Breakeven point = Strike Price + Premium

Buying a Put Option:Profit increases as the spot price decreases below the strike price (up to K - P when S=0)
                    Maximum loss is limited to the premium paid
                    Breakeven point = Strike Price - Premium

Selling a Put Option:Profit is limited to the premium received
                     Loss increases as the spot price decreases below the strike price (up to K - P when S=0)
                     Breakeven point : The seller profits if S stays above K (keeping P). If S drops below K, losses increase. The point where losses offset the premium is 
                     S=KP.
```
